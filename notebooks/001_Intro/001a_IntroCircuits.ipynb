```python
from qiskit import QuantumCircuit
from qiskit.providers.aer import AerSimulator
```

```python
# Create quantum circuit of 3 qubits
qc = QuantumCircuit(3,3)
# Measure each qubit with his classical bit
qc.measure([0,1,2],[0,1,2])
# Draw the circuit in a certain style
qc.draw(output='mpl')
```

```python
# Create a simulator
sim = AerSimulator()
job    = sim.run(qc)  # run experimetn
result = job.result() # get the results
result.get_counts()   # interpret the results as a counts dictionary
```

# CNOT Gate

```python
qc = QuantumCircuit(2,2)
qc.x(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])
display(qc.draw(output='mpl'))

sim = AerSimulator()
job = sim.run(qc)
result = job.result()
print(f'Results: {result.get_counts()}')
```

# Half adder algorithm
0+0=00

0+1=01

1+0=01

1+1=10

```python
qc = QuantumCircuit(4,2)

# Encode an input (here '11')
qc.x(0)
qc.x(1)

# Carry out the adder circuit we created
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,3)

# Measure the bottom two qubits to obtain the output
qc.measure(2,0)
qc.measure(3,1)

# draw
qc.draw(output='mpl')
```

```python
job = sim.run(qc)  # run the experiment
result = job.result()   # get the results
result.get_counts()     # interpret the results as a “counts” dictionary
```
