### A 3-SAT problem:

```python
with open('../data/3example.dimacs', 'r', encoding='utf8') as f:
    dimacs = f.read()
print(dimacs)  # let's check the file is as promised
```

Now we can use Qiskit to transform this to an oracle:

```python
from qiskit.circuit.library import PhaseOracle, GroverOperator
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram, array_to_latex
import numpy as np
```

```python
oracle = PhaseOracle.from_dimacs_file('../data/3example.dimacs')
oracle.draw(output='mpl')
```

```python
init = QuantumCircuit(3)
init.h([0,1,2])
init.draw(output='mpl')
```

```python
grover_op = GroverOperator(oracle)

# combine the circuits
qc = init.compose(grover_op)
qc.measure_all()
qc.draw(output='mpl')
```

Let's run it on a simulator:

```python
# simulate the circuit
sim    = Aer.get_backend('aer_simulator')
t_qc   = transpile(qc, sim)
counts = sim.run(t_qc).result().get_counts()

# plot the results
print('We have high probability of measure the 3 solutions of the SAT problem')
plot_histogram(counts)
```

# Simple circuit for Grover's algorithm

### Target the 11 solution:

```python
oracle = QuantumCircuit(2)
oracle.cz(0,1) # invert phase of |11>
oracle.draw(output='mpl')
```

```python
def display_unitary(qc, prefix=""):
    """Simulates a simple circuit and display its matrix representation.
    Args:
        qc (QuantumCircuit): The circuit to compile to a unitary matrix
        prefix (str): Optional LaTeX to be displayed before the matrix
    Returns:
        None (displays matrix as side effect)
    """
    sim = Aer.get_backend('aer_simulator')
    # Next, we'll create a copy of the circuit and work on
    # that so we don't change anything as a side effect
    qc = qc.copy()
    # Tell the simulator to save the unitary matrix of this circuit
    qc.save_unitary()
    unitary = sim.run(qc).result().get_unitary()
    display(array_to_latex(unitary, prefix=prefix))

display_unitary(oracle, "U_\\text{oracle}=")
```

```python
diffuser = QuantumCircuit(2)
diffuser.h([0, 1])
diffuser.draw(output='mpl')
```

So, applying the X-gate to each qubit, will do the transformation 00 -> 11:

```python
diffuser.x([0,1])
diffuser.draw(output='mpl')
```

Now, we have the transformation s -> 11, we can apply our CZ-gate and reverse the transformation:

```python
diffuser.cz(0,1)
diffuser.x([0,1])
diffuser.h([0,1])
diffuser.draw(output='mpl')
```

```python
grover = QuantumCircuit(2)
grover.h([0,1])
grover.barrier()
grover = grover.compose(oracle)
grover.barrier()
grover = grover.compose(diffuser)
grover.measure_all()
grover.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
counts = sim.run(grover).result().get_counts()
print(counts)
plot_histogram(counts)
```

### Target 00

```python
oracle = QuantumCircuit(2)
# invert phase og 00
oracle.z([0,1])
oracle.cz(0,1)
oracle.rx(2*np.pi,0)
oracle.draw(output='mpl')
```

```python
display_unitary(oracle, "U_\\text{oracle}=")
```

```python
diffuser = QuantumCircuit(2)
diffuser.h([0,1])
diffuser.draw(output='mpl')
```

Now we need to transform 00 to 00, so we do nothing. We only need transform s to 00, so we apply our transformation

```python
diffuser.z([0,1])
diffuser.cz(0,1)
diffuser.rx(2*np.pi,0)
diffuser.h([0,1])
diffuser.draw(output='mpl')
```

Put together:

```python
grover = QuantumCircuit(2)
grover.h([0,1]) # initialize s
grover.barrier()
grover = grover.compose(oracle)
grover.barrier()
grover = grover.compose(diffuser)
grover.measure_all()
grover.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
counts = sim.run(grover).result().get_counts()
print(counts)
plot_histogram(counts)
```
